```python
pi = np.pi; nan = np.nan
df_dead = df.query('GISRLat > @pi')
df_dead.info()
df_dead.head()
```

```python
# bounding rectangles
map_grid = {}
limit = 0
i = 0
for index, row in df_dead.iterrows():
    grid_name = '{}_{}'.format(int(row['LAT_DECIMAL_DEGREES']),int(row['LON_DECIMAL_DEGREES'])) 
    if not grid_name in map_grid:
        map_grid[grid_name] = 1
    else:
        map_grid[grid_name] += 1
        
    if limit > 0:
        i += 1    
    if limit > 0 and i > limit:
        break
    
print('map_grid: ', map_grid) 

print('map grids with ducks: ', len(map_grid))
```
